```python
import pandas as pd
import csv
from function import map_month, get_season
```

```python
# shows us what were working with
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"
shark_data =  pd.read_excel(url)  # usecols=["Embarked"]
shark_data.head()
```

```python
# replace all sentence and whitespace from the string date before handling to season
shark_data['Date'] = shark_data['Date'].replace('Before ', '', regex=True)
shark_data['Date'] = shark_data['Date'].replace(' ', '-', regex=True)
# Extract all months ans year from the columns date
shark_data['Month'] = shark_data['Date'].str.extract(r'([a-zA-Z]+)')
shark_data['Year']= shark_data['Date'].str.extract(r'(\d{4})')

shark_data['Month'] = shark_data['Month'].apply(map_month)
# Finish the category seasson for the months
shark_data['Season'] = shark_data['Month'].apply(get_season)
```

```python
url = "https://marineprotectedareas.noaa.gov/media/data/NOAA_Marine_Protected_Areas_Inventory_2023.xlsx"
usa_shark_data =  pd.read_excel(url)  
usa_shark_data.head()
```

```python
shark_data.dropna(thresh=6, inplace=True) # to delete rows with 5 or less empty values
```

```python
# drops all unwanted columns and saves the results to a series called shark_data_trimmed
drop_cols = ['pdf', 'href formula', 'href', 'Case Number', 'Case Number.1', 'Species ', 'original order', 'Unnamed: 21', 'Unnamed: 22', 'Source', 'Unnamed: 11', "Name"]
shark_data_trimmed = shark_data.drop( drop_cols, axis=1)

shark_data_trimmed

# Note: think about doing this step before the provoked and unprovoked variables so its based off the trimmed data
```

```python
usa_attack_data = shark_data_trimmed.groupby('Country').get_group('USA')
usa_attack_data
```

```python
# makes a series variable of data for provoked attacks
provoked_data = usa_attack_data[usa_attack_data['Type'] == 'Provoked']
provoked_data_unique = provoked_data.drop_duplicates(subset=['Activity'], keep='first')
provoked_data
```

```python
# makes a series variable of data for unprovoked attacks
non_provoked_data = usa_attack_data[usa_attack_data['Type'] != 'Provoked']
non_provoked_data
```
