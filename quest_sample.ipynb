```python
"""Hypothesis 3: Season and Frequency of Attacks: While we do not have specific data on weather events,
we can explore whether certain times of year have a higher incidence of shark attacks by using the "Date" 
column to identify the season in which the attacks occurred.""" 

import pandas as pd
import matplotlib.pyplot as plt
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"

shark_data = pd.read_excel(url)

# Remove all columns irrelevant for this analysis 

drop_cols = ['pdf', 'hrefformula', 'href', 'CaseNumber', 'CaseNumber.1', 'originalorder', 'Unnamed:21', 'Unnamed:22', 'Source', 'Species',  'Unnamed:11',	"Name", "Location"]
shark_data.columns = shark_data.columns.str.replace(' ', '')
shark_data.drop(drop_cols, axis=1, inplace=True)
# Remove all countries that are irrelevant for this analysis 
shark_data.drop(shark_data[shark_data['Country'] != 'USA'].index, inplace=True)
```

```python
shark_data.isnull().sum()
```

```python
# replace all sentence that are not usefull as a filter and replaced to Injured to compare to seasson
shark_data.loc[shark_data['Injury'] != 'Fatal', 'Injury'] =  "Injured"
```

```python
# Conveting Age into numerical and creating a Category.
shark_data['Age'] = pd.to_numeric(shark_data['Age'], errors='coerce')
shark_data.loc[shark_data['Age'] < 18, 'Category'] = 'Child'
shark_data.loc[(shark_data['Age'] >= 18) & (shark_data['Age'] <= 30), 'Category'] = 'Young'
shark_data.loc[shark_data['Age'] > 30, 'Category'] = 'Adult'
```

```python
# replace all sentence and whitespace from the string date before handling to season

shark_data['Date'] = shark_data['Date'].replace('Before ', '', regex=True)
shark_data['Date'] = shark_data['Date'].replace(' ', '-', regex=True)

# Extract all months ans year from the columns date 


shark_data['Month'] = shark_data['Date'].str.extract(r'([a-zA-Z]+)')
shark_data['Year']= shark_data['Date'].str.extract(r'(\d{4})')

# Create a Category for all months 
def map_month(month_abbr):
    month_abbr_to_full = {
        'Jan': 'January', 'Feb': 'February', 'Mar': 'March', 'Apr': 'April',
        'May': 'May', 'Jun': 'June', 'Jul': 'July', 'Aug': 'August',
        'Sep': 'September', 'Oct': 'October', 'Nov': 'November', 'Dec': 'December'
    }
    return month_abbr_to_full.get(month_abbr, 'Unknown')

def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Fall'
    else:
        return 'Unknown'
    
shark_data['Month'] = shark_data['Month'].apply(map_month)
# Finish the category seasson for the months 
shark_data['Season'] = shark_data['Month'].apply(get_season)

```
